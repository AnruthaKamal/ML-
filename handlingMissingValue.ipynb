## Handling Missing Values 

```python
df.isnull().sum()
```

20 missing values in the following columns : CRIM , ZN , INDUS , CHAS , AGE , LSTAT 

```python
sns.distplot(df['CRIM'])
```

```python
fig, axes = plt.subplots(2, 3,figsize=(10,10))

#create chart in each subplot
sns.distplot(df['CRIM'], ax=axes[0,0])
sns.distplot(df['ZN'], ax=axes[0,1])
sns.distplot(df['INDUS'], ax=axes[0,2])
sns.distplot(df['CHAS'], ax=axes[1,0])
sns.distplot(df['AGE'], ax=axes[1,1])
sns.distplot(df['LSTAT'], ax=axes[1,2])
```

```python
fig, axes = plt.subplots(2, 3,figsize=(10,10))

#create chart in each subplot
sns.boxplot(df['CRIM'], ax=axes[0,0])
sns.boxplot(df['ZN'], ax=axes[0,1])
sns.boxplot(df['INDUS'], ax=axes[0,2])
sns.boxplot(df['CHAS'], ax=axes[1,0])
sns.boxplot(df['AGE'], ax=axes[1,1])
sns.boxplot(df['LSTAT'], ax=axes[1,2])
```

 <ul>

 <li>CRIM , ZN , LSTAT has outliers hence missing values in these can be imputed by <b>median</b></li> 

 <li>Missing Values is AGE and INDUS can be imputed by <b>mean</b> </li>

 <li>Missing values in CHAS can be imputed by <b>mode</b></li>

```python
df['CRIM'].fillna(df['CRIM'].median(),inplace = True)
df['ZN'].fillna(df['CRIM'].median(),inplace = True)
df['INDUS'].fillna(df['INDUS'].mean(),inplace = True)
df['CHAS'].fillna(df['CHAS'].mode(),inplace = True)
df['AGE'].fillna(df['AGE'].mean(),inplace = True)
df['LSTAT'].fillna(df['LSTAT'].median(),inplace = True)
```
